# Mp3 Project

```octave
fileName='wavs/8k8bitpcm.wav';
[y, fs]=wavread(fileName);
```

```octave
interval = 1:5000;
```

```octave
time = (1:length(y(interval))) / fs;
```

## Quantization

```octave
plot(time,y(interval))
```

```octave
plot(time, filterAndQuantize(y(interval)))
```

## Huffman Encoding

```octave
[encodedFre, fres, huffTable] = huffEncodeFre(y(interval)');
disp('Not quantized huffman')
disp(length(encodedFre));
```

```octave
disp('Quantized huffman')
[encodedFre, fres, huffTable] = huffEncodeFre(filterAndQuantize(y(interval))');
disp(length(encodedFre));
```

```octave
disp('Real File size')
disp(length(y(interval)) * 17);
```

## Decoding

```octave
fileName='wavs/8k8bitpcm.wav';
[y, fs]=wavread(fileName);
```

```octave
interval = 1:200;
time = (1:length(y(interval))) / fs;
plot(time, y(interval));
[encodedFre, fres, huffTable] = huffEncodeFre(y(interval)');
```

```octave
decodeY = huffDecodeFre(encodedFre, fres, huffTable);
plot(time, decodeY);
```

## Lossy Decoding 

```octave
[encodedFre, fres, huffTable] = huffEncodeFre(filterAndQuantize(y(interval))');
decodeY = huffDecodeFre(encodedFre, fres, huffTable);
plot(time, decodeY);
```
